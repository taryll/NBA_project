```python
from nba_api.stats.endpoints import leaguegamefinder
```

```python
def find_game_df(season_nullable):
#     print(season_nullable)
    league_game = leaguegamefinder.LeagueGameFinder(league_id_nullable = "00", 
                                                    season_type_nullable = "Regular Season",
                                                    season_nullable = season_nullable)

    df_nba_games = league_game.league_game_finder_results.get_data_frame()
    return df_nba_games

def create_df_nba_games(df_nba_games):
    # separate between home or away games
    df_nba_games['WHERE'] = df_nba_games.apply(lambda x: "Away" if "@" in x['MATCHUP'] else "Home", axis = 1)

    df_home_games = df_nba_games[df_nba_games['WHERE'] == "Home"]
    df_away_games = df_nba_games[df_nba_games['WHERE'] == "Away"]
    
    df_home_games = df_home_games.rename(columns = dict_home_columns)
    df_away_games = df_away_games.rename(columns = dict_away_columns)
    
    df_game_merge = pd.merge(df_home_games, df_away_games, how = "inner", left_on = "H_GAME_ID", right_on = "A_GAME_ID")
    
    return df_game_merge
```

```python
season_list = ["2021-22", "2020-21", "2019-20", "2018-19", "2017-18", "2016-17", "2015-16", 
               "2014-15", "2013-14", "2012-13", "2011-12", "2010-11", "2009-10"]

df_nba_games = pd.concat(list(map(lambda x: create_df_nba_games(find_game_df(x)), season_list)))
```

# Web Scrapping

To collect player rating on NBA 2k, we will need to use web scrap on hoopshype.com

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

from selenium import webdriver
import time
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.114 Safari/537.36'}

season_list = ["2021-2022", "2020-2021", "2019-2020", "2018-2019", "2017-2018", "2016-2017", "2015-2016", 
               "2014-2015", "2013-2014", "2012-2013", "2011-2012", "2010-2011", "2009-2010"]

path = "../data/raw/"

for season in season_list:    
    url = ("https://hoopshype.com/nba2k/" + season)
    response = requests.get(url, headers = headers)
    html = response.text
    soup = BeautifulSoup(html, "html.parser")
    
    player_rating = soup.find("table", attrs = {"class": "hh-salaries-ranking-table hh-salaries-table-sortable responsive"})
    player_rating = player_rating.find("tbody")
    player_rating = player_rating.find_all("tr")
    
    nba_2k = []

    for player in player_rating:
#         print(player.find("td", attrs = {"class": "name"}).get_text)
        nba_dict = {}
        nba_dict["name"] = player.find("td", attrs = {"class": "name"}).get_text(strip = True)
        nba_dict["rating"] = player.find("td", attrs = {"class": "value"}).get_text(strip = True)

        nba_2k.append(nba_dict)
    
    print("nba 2k season " + season + " completed")
    
    df_player_rating = pd.DataFrame(nba_2k)
    df_player_rating['season'] = season[:5] + season[-2:]
    df_path = path + "df_player_rating_2k" + season[-2:] + ".csv"
    df_player_rating.to_csv(df_path, index = False)
```
