```python
import pandas as pd
import numpy as np
import janitor
```

```python
df_path = ("../data/raw/")
```

```python
df_player_stats = pd.read_csv(df_path + "df_player_stats.csv")
```

```python
df_nba_teams = pd.read_csv(df_path + "df_nba_teams.csv")
```

```python
# we need to join player rating by name
df_nba_players = pd.read_csv("../data/raw/df_nba_players.csv")
# df_nba_players.head()

df_nba_players = df_nba_players.loc[:, ['id', 'full_name']]
df = pd.merge(df_player_stats, df_nba_players, how = "left", left_on = "PLAYER_ID", right_on = "id")
```

NJN stands for New Jerseys Nets (Brooklyn Nets)

NOH stands for New Orleans Hornets (New Orleans Pelicans)

NOK stands for New Orleans/Oklahoma City Hornets (New Orleans Pelicans)

SEA stands for Seattle SuperSonics (Oklahoma City Thunder)

TOT stands for Two Other Teams

```python
nba_teams_dict = {"NJN": "BKN",
                  "NOH": "NOP",
                  "NOK": "NOP",
                  "SEA": "OKC"}
```

```python
df = df.replace({'TEAM_ABBREVIATION': nba_teams_dict})
```

```python
# remove TOT
df = df[df['TEAM_ABBREVIATION'] != "TOT"]
```

```python
df.groupby(['full_name', 'SEASON_ID', 'PLAYER_AGE']).sum()
```

```python
# some players played in more than one team in season, so 
# groupby their name and season and sum(), actually
# where they played doesnt matter now

df_group = df.groupby(['full_name', 'SEASON_ID', 'PLAYER_AGE']).sum()
df_group = df_group.reset_index()
```

```python
df_group
```

```python
valid_season_list = ["2021-22", "2020-21", "2019-20", "2018-19", "2017-18", "2016-17", "2015-16", 
                     "2014-15", "2013-14", "2012-13", "2011-12", "2010-11", "2009-10"]

df_temp_list = []
columns_filt = ['SEASON_ID', 'full_name']
df_filt = df_group.loc[:, columns_filt]
soma_lista = []

for season in valid_season_list:
    df_left = df_filt[df_filt['SEASON_ID'] == season]
    df_right = pd.read_csv(df_path + "df_player_rating_2k" + season[-2:] + ".csv")
#     print(df_left.shape[0])
    df_left.shape[0]
    df_temp = pd.merge(df_left, df_right, how = "left", left_on = "full_name", right_on = "name")
    df_temp_list.append(df_temp)
# df_temp_list
```

```python
df_player_rating = pd.concat(df_temp_list).loc[:, ['SEASON_ID', 'full_name', 'rating']]
```

```python
unvalid_season_list = list(set.difference(set(df_group['SEASON_ID'].unique()),set(valid_season_list)))
df_temp = df_filt[df_filt['SEASON_ID'].isin(unvalid_season_list)]
df_temp.loc[:, 'rating'] = np.NaN

df_temp
```

```python
df_player_rating = pd.concat([df_player_rating, df_temp])
```

```python
df_player_rating_stats = pd.merge(df_group, df_player_rating, how = 'inner', 
                                  left_on = ['full_name', 'SEASON_ID'],
                                  right_on = ['full_name', 'SEASON_ID']).drop_duplicates()
```

```python
# create useful features like ppg, rpg
df_player_rating_stats['PTSPG'] = df_player_rating_stats['PTS'] / df_player_rating_stats['GP']
df_player_rating_stats['REBPG'] = df_player_rating_stats['REB'] / df_player_rating_stats['GP']
df_player_rating_stats['OREBPG'] = df_player_rating_stats['OREB'] / df_player_rating_stats['GP']
df_player_rating_stats['DREBPG'] = df_player_rating_stats['DREB'] / df_player_rating_stats['GP']
df_player_rating_stats['ASTPG'] = df_player_rating_stats['AST'] / df_player_rating_stats['GP']
df_player_rating_stats['STLPG'] = df_player_rating_stats['STL'] / df_player_rating_stats['GP']
df_player_rating_stats['BLKPG'] = df_player_rating_stats['BLK'] / df_player_rating_stats['GP']
df_player_rating_stats['TOVPG'] = df_player_rating_stats['TOV'] / df_player_rating_stats['GP']
df_player_rating_stats['PFPG'] = df_player_rating_stats['PF'] / df_player_rating_stats['GP']
# df_player_stats['RPG'] = df_player_stats['PTS'] / df_player_stats['GP']
```

```python
# remaking features affected by groupby
df_player_rating_stats['FG_PCT'] = df_player_rating_stats['FGM'] / df_player_rating_stats['FGA']
df_player_rating_stats['FG3_PCT'] = df_player_rating_stats['FG3M'] / df_player_rating_stats['FG3A']
df_player_rating_stats['FT_PCT'] = df_player_rating_stats['FTM'] / df_player_rating_stats['FTA']
```

```python
df_player_rating_stats = df_player_rating_stats.clean_names()
```
