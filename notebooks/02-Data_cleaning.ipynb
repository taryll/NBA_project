```python
def create_df_nba_games(df_nba_games):
    # separate between home or away games
    df_nba_games['WHERE'] = df_nba_games.apply(lambda x: "Away" if "@" in x['MATCHUP'] else "Home", axis = 1)

    df_home_games = df_nba_games[df_nba_games['WHERE'] == "Home"]
    df_away_games = df_nba_games[df_nba_games['WHERE'] == "Away"]
    
    nba_columns = df_nba_games.columns.to_list()
    home_columns = list(map(lambda x: "H_" + x, nba_columns))
    away_columns = list(map(lambda x: "A_" + x, nba_columns))
    
    dict_home_columns = dict(zip(nba_columns, home_columns))
    dict_away_columns = dict(zip(nba_columns, away_columns))
    
    df_home_games = df_home_games.rename(columns = dict_home_columns)
    df_away_games = df_away_games.rename(columns = dict_away_columns)
    
#     df_game_merge = pd.merge(df_home_games, df_away_games, how = "inner", left_on = "H_GAME_ID", right_on = "A_GAME_ID")
    
    return df_home_games, df_away_games
```

### Elo Rating

$ R_0 = 1500 $ start value

$ R_{i+1} = K(S_{team} - E_{team}) + R_i$

$S_{team}$: boolean for win or lose (1 or zero)

Expected win probability:

$E_{team} = \dfrac{1}{1 + 10^{\left(\dfrac{opp\_elo - team\_elo}{400}\right )} }$

$K = 20 \dfrac{\left(MOV_{winner} + 3\right)^{0.8}}{7.5 + 0.006(elo\_difference_{winner})}$

for new season (elo rating carry over from season to season)

$ (R * 0.75) + 0.25*1505$

```python
def s_team(home_score, away_score):
    if home_score > away_score:
        home_s = 1
        away_s = 0
    else:
        home_s = 0
        away_s = 1
    
    return home_s, away_s

def exp_win_prob(opp_elo, team_elo):
    denom = 1 + 10**((opp_elo - team_elo) / 400)
    
    return 1 / denom

def k_fact(mov, elo_diff):
    c = 20

    # means home wins
    if mov > 0 :
        value = ((mov + 3)**(0.8)) / (7.5 + 0.006*(elo_diff))
    # means home loses
    else:
        value = ((-mov + 3)**(0.8)) / (7.5 + 0.006*(-elo_diff))
        
    return c*value

def elo_update(home_elo, away_elo, home_score, away_score, home_court_adv = 100):
    # home advantage
    home_elo += home_court_adv
    
    home_s, away_s = s_team(home_score, away_score)
    
    home_prob = exp_win_prob(away_elo, home_elo)
    away_prob = exp_win_prob(home_elo, away_elo)
    
    k = k_fact(home_score - away_score, home_elo - away_elo)
    
    # home advantage is not for update
    home_elo -= home_court_adv
    home_elo_update = k*(home_s - home_prob) + home_elo
    away_elo_update = k*(away_s - away_prob) + away_elo
    
    return home_elo_update, away_elo_update

def get_prev_elo(df, date, team_id, season, df_elo):
    df_temp = df[df['GAME_DATE'] < date][(df['H_TEAM_ID'] == team_id) | (df['A_TEAM_ID'] == team_id)]
    prev_game = df_temp.sort_values(by = 'GAME_DATE').tail(1).iloc[0]
    
    if team_id == prev_game['H_TEAM_ID']:
        team_elo = df_elo[df_elo['GAME_ID'] == prev_game['GAME_ID']]['H_ELO_AFTER'].values[0]
    else :
        team_elo = df_elo[df_elo['GAME_ID'] == prev_game['GAME_ID']]['A_ELO_AFTER'].values[0]
        
    if prev_game['SEASON_ID'] != season :
        return (0.75 * team_elo) + (0.25 * 1505)
    else :
        return team_elo
```

```python
filt_columns = ['H_SEASON_ID', 'H_TEAM_ID', 'H_TEAM_NAME', 'A_TEAM_ID', 'A_TEAM_NAME',
                'GAME_ID', 'GAME_DATE', 'H_WL', 'A_WL', 'H_PTS', 'A_PTS']

df_nba_filt = df_nba_merg.loc[:, filt_columns]
df_nba_filt.rename(columns = {'H_GAME_DATE': 'GAME_DATE',
                              'H_GAME_ID': 'GAME_ID',
                              'H_SEASON_ID': 'SEASON_ID'}, 
                   inplace = True)

df_nba_filt.sort_values(by = 'GAME_DATE', inplace = True)
df_nba_filt.reset_index(drop = True, inplace = True)
```

```python
df_elo = pd.DataFrame(columns = ['GAME_ID', 'H_TEAM_ID', 'A_TEAM_ID',
                                 'H_ELO_BEFORE', 'A_ELO_BEFORE',
                                 'H_ELO_AFTER', 'A_ELO_AFTER'])

df_teams_elo = pd.DataFrame(columns = ['GAME_ID', 'TEAM', 'ELO', 'DATE', 'WHERE_PLAYED', 'SEASON']) 

for index, row in df_nba_filt.iterrows():
    date = row['GAME_DATE']
    game_id = row['GAME_ID']
    season = row['SEASON_ID']
    
    home_team_id, home_score = row['H_TEAM_ID'], row['H_PTS']
    away_team_id, away_score = row['A_TEAM_ID'], row['A_PTS']
    
    if (home_team_id not in df_elo['H_TEAM_ID'].values and home_team_id not in df_elo['A_TEAM_ID'].values):
        home_elo_before = 1500
    else:
        home_elo_before = get_prev_elo(df_nba_filt, date, home_team_id, season, df_elo)

    if (away_team_id not in df_elo['H_TEAM_ID'].values and away_team_id not in df_elo['A_TEAM_ID'].values):
        away_elo_before = 1500
    else:
        away_elo_before = get_prev_elo(df_nba_filt, date, away_team_id, season, df_elo)

    home_elo_after, away_elo_after = elo_update(home_elo_before, away_elo_before,
                                                home_score, away_score, 100)
    
    new_row = {'GAME_ID': game_id,
               'H_TEAM_ID': home_team_id,
               'A_TEAM_ID': away_team_id,
               'H_ELO_BEFORE': home_elo_before,
               'A_ELO_BEFORE': away_elo_before,
               'H_ELO_AFTER': home_elo_after,
               'A_ELO_AFTER': away_elo_after}
    
    teams_row_one = {'GAME_ID': game_id,
                     'TEAM': home_team_id, 
                     'ELO': home_elo_before,
                     'DATE': date,
                     'WHERE_PLAYED': 'Home',
                     'SEASON': season}
    
    teams_row_two = {'GAME_ID': game_id,
                     'TEAM': away_team_id,
                     'ELO': away_elo_before,
                     'DATE': date,
                     'WHERE_PLAYED': 'Away',
                     'SEASON': season}
    
    df_elo = df_elo.append(new_row, ignore_index = True)
    
    df_teams_elo = df_teams_elo.append(teams_row_one, ignore_index=True)
    df_teams_elo = df_teams_elo.append(teams_row_two, ignore_index=True)

    
```

```python
elo_columns = ['GAME_ID', 'H_ELO_BEFORE', 'A_ELO_BEFORE', 'H_ELO_AFTER', 'A_ELO_AFTER']

df_nba_with_elo = pd.merge(df_nba_merg, df_elo.loc[:, elo_columns], how = "left", 
                           left_on = 'GAME_ID', right_on = 'GAME_ID')
```

### Team Performance last n games

```python
def get_avg_stats_last_n_games(df, team_id, date, n):
    df_temp = df[df['GAME_DATE'] < date][(df['H_TEAM_ID'] == team_id) | (df['A_TEAM_ID'] == team_id)]
    prev_game = df_temp.sort_values(by = 'GAME_DATE').tail(n)

    cols = ['TEAM_ID', 'PTS', 'FGM', 'FGA', 'FG_PCT', 'FG3M', 'FG3A',
            'FG3_PCT', 'FTM', 'FTA', 'FT_PCT', 'OREB', 'DREB',
            'REB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PLUS_MINUS']
    
    home_cols = ["H_" + x for x in cols]
    away_cols = ["A_" + x for x in cols]
    
    df_home = prev_game.loc[:, home_cols]
    df_away = prev_game.loc[:, away_cols]
    
    df_home.columns = [x[2:] for x in df_home.columns]
    df_away.columns = [x[2:] for x in df_away.columns]
    
    df_team = pd.concat([df_home, df_away])
    df_team = df_team[df_team['TEAM_ID'] == team_id]
    df_team.drop(columns = ['TEAM_ID'], inplace=True)

    return df_team.mean()
```

```python
season = 22021
df_temp = df_nba_merg[df_nba_merg['H_SEASON_ID'] == season]
df_season_games = df_temp.sort_values(by = 'GAME_DATE').reset_index(drop = True)

df_season_recent_perf = pd.DataFrame()
i = 0
for index, row in df_season_games.iterrows():
    date = row['GAME_DATE']
    game_id = row['GAME_ID']
    home_team_id = row['H_TEAM_ID']
    away_team_id = row['A_TEAM_ID']
    
    h_team_recent_performance = get_avg_stats_last_n_games(df_season_games, home_team_id, date, 10)
    h_team_recent_performance.index = ['H_Last_10_Avg_' + x for x in h_team_recent_performance.index]

    a_team_recent_performance = get_avg_stats_last_n_games(df_season_games, away_team_id, date, 10)
    a_team_recent_performance.index = ['A_Last_10_Avg_' + x for x in a_team_recent_performance.index]

    new_row = pd.concat([h_team_recent_performance, a_team_recent_performance], sort = False)
    new_row['GAME_ID'] = game_id
    
    df_season_recent_perf = df_season_recent_perf.append(new_row, ignore_index = True)
    df_season_recent_perf = df_season_recent_perf[new_row.index]

df_recent_perf = pd.concat([df_recent_perf, df_season_recent_perf])
```
